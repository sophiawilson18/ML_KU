### Method 2: Using sklearn.feature_selection.SelectFromModel

Procedure
1. SelectFromModel determines the 25 best features.
2. I remove the top 10 most correlated features since SelectFromModel does not take this into account. 

Notes:
- All 15 features found using the Gini importance is also found using SelectFromModel

```python
from sklearn.feature_selection import SelectFromModel
rf = RandomForestClassifier()
sel = SelectFromModel(rf, max_features=25)
sel.fit(X_train, y_train)
```

```python
sel.get_support()
features2 = X_train.columns[(sel.get_support())]
print('Number of selected features', len(features2))
#print('Selected features', features2)
```

```python
N = 0
for i in features1:
    if i in features2: N+=1
    
print('Number of overlapping features in the two methods:', N)
```

```python
def binary_split(X, y, features):
    '''Splits data into positive and negative'''
    df_X = pd.DataFrame(X[features]) 
    df_1 = df_X.loc[y==1]
    df_0 = df_X.loc[y==0]
    return df_X, df_1, df_0

def corr_plot(df, ax, title):
    '''Correlation plot'''
    sns.heatmap(df.corr(), mask=np.zeros_like(df.corr(), dtype=bool), cbar_kws={"shrink": 0.65},
            cmap=sns.diverging_palette(220, 10, as_cmap=True), vmin=-1.0, vmax=1.0, square=True, ax=ax)
    ax.set_title(title)

def drop_high_corr(df, cut=0.95):
    '''Drops features with Pearson correlation > cut'''
    cor_matrix = df.corr(method='pearson').abs() 
    upper_tri = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(bool)) 
    to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > cut)] 
    df_new = df.drop(df[to_drop], axis=1) 
    features_new = df_new.keys()
    return df_new, features_new
```

```python
# Divide data into electrons and non-electrons:
df_X, df_1, df_0 = binary_split(X, y, features2)

# Make two separate correlation plots, also to see differences:
fig, ax = plt.subplots(1, 2, figsize=(14, 8))
corr_plot(df_1, ax[0], 'electrons')
corr_plot(df_0, ax[1], 'non electrons')
fig.tight_layout();
```

```python
_, features_new = drop_high_corr(df_X, cut=0.57)
_, df_1_new, df_0_new = binary_split(df_X, y, features_new)

# Make two separate correlation plots, also to see differences:
fig, ax = plt.subplots(1, 2, figsize=(14, 8))
corr_plot(df_1_new, ax[0], 'electrons')
corr_plot(df_0_new, ax[1], 'non electrons')
fig.tight_layout();
```
